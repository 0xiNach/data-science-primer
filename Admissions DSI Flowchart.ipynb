```python
from graphviz import Digraph

# See: http://www.graphviz.org/doc/info/attrs.html


galvanize_orange = '#F37D19'
galvanize_gray = '#96989C'


def start_end(g, name, label=None, line_color='#444444', fill_color='#F37D19', font_color='#FFFFFF', pen_width=3):
    if label is None:
        label = name
    g.node(name, label=label, shape='box', style='rounded, filled', color=line_color,
                                           fillcolor=fill_color, fontcolor=font_color,
                                           penwidth=str(pen_width))


def condition(g, name, label=None, line_color='#444444', fill_color='#FFFFFF', font_color='#F37D19', pen_width=3):
    if label is None:
        label = name
    g.node(name, label=label, shape='diamond', style='rounded, filled', color=line_color,
                                               fillcolor=fill_color, fontcolor=font_color,
                                               penwidth=str(pen_width))


def action(g, name, label=None, line_color='#F37D19', fill_color='#FFFFFF', font_color='#444444', pen_width=3, url=None):
    if label is None:
        label = name
    g.node(name, label=label, shape='parallelogram', style='rounded, filled', color=line_color,
                                                     fillcolor=fill_color, fontcolor=font_color,
                                                     penwidth=str(pen_width), URL=url)


def checkpoint(g, name, label=None, line_color='#444444', fill_color='#F37D19', font_color='#FFFFFF', pen_width=3):
    if label is None:
        label = name
    g.node(name, label=label, shape='circle', style='filled', color=line_color,
                                              fillcolor=fill_color, fontcolor=font_color,
                                              penwidth=str(pen_width))
```

```python
g = Digraph(format='pdf', engine='dot')   # dot (default), circo, neato, fdp, sfdp, twopi

start_end(g, 'start')
start_end(g, 'apply', 'Schedule Python Interview!')

condition(g, 'programming', 'Programming\nExperience?')
condition(g, 'python', 'Python\nExperience?')

action(g, 'hardway', 'Learn Python\nthe Hard Way\n\ngoo.gl/exWE1w')
action(g, 'googleclass', 'Google Python\nCourse\n\ngoo.gl/OejaOz')
action(g, 'codewars', 'Practice Python\non CodeWars\n\ngoo.gl/ANAdMA')

condition(g, 'level6', 'Kata Level 6\non CodeWars?\n\n(Level 5 would\nbe better!)\n\ngoo.gl/4qASmG')

action(g, 'practice1', 'Practice!')
action(g, 'practice2', 'Practice\nmore!')

g.edge('start', 'programming')

g.edge('programming', 'python', label='yes')
g.edge('programming', 'hardway', label='no')
g.edge('hardway', 'googleclass')
g.edge('googleclass', 'codewars')

g.edge('python', 'codewars', label='yes')
g.edge('python', 'googleclass', label='no')

g.edge('codewars', 'level6')

g.edge('level6', 'apply', label='yes')
g.edge('level6', 'practice1', label='no')

g.edge('practice1', 'practice2')
g.edge('practice2', 'level6')

g.render('programming')

g
```

```python
g = Digraph(format='pdf', engine='dot')   # dot (default), circo, neato, fdp, sfdp, twopi

start_end(g, 'start')
start_end(g, 'apply', "Schedule Stats Interview!")

condition(g, 'concepts', "New to\nstatistical CONCEPTS,\nor don't remember\nmuch?")
condition(g, 'mathematical', "New to\nMATHEMATICAL statistics,\nor don't remember\nmuch?")
condition(g, 'conditional_prob', "Can you solve\nconditional probability\nword problems?")
condition(g, 'distributions', "Can you model\na problem using well-\nknown distributions?")
condition(g, 'regressionML', "Do you know\nthe foundational linear\nregression and\nmachine learning\nconcepts?")

action(g, 'study_concepts', 'Study basic\nstatistical concepts\n\ngoo.gl/wABcIh')
action(g, 'study_math', 'Study how to represent\nstatistical concepts\nin mathematics\n\ngoo.gl/NYSybO')
action(g, 'study_conditional_prob', 'Review conditional\nprobability\n\ngoo.gl/wpd9Us')
action(g, 'study_distributions', 'Review the well-known\ndistributions\n(unifrom, normal, Poisson,\nBernoulli, etc.)\n\ngoo.gl/p4eQmR')
action(g, 'study_regressionML', 'Watch or read\nchapters 1-4 of\nAn Intro to\nStatistical Learning\n\ngoo.gl/KyzCRE')

g.edge('start', 'concepts')

g.edge('concepts', 'study_concepts', label='yes')
g.edge('concepts', 'mathematical', label='no')

g.edge('mathematical', 'study_math', label='yes')
g.edge('mathematical', 'conditional_prob', label='no')

g.edge('conditional_prob', 'study_conditional_prob', label='no')
g.edge('conditional_prob', 'distributions', label='yes')

g.edge('distributions', 'study_distributions', label='no')
g.edge('distributions', 'regressionML', label='yes')

g.edge('regressionML', 'study_regressionML', label='no')
g.edge('regressionML', 'apply', label='yes')

g.edge('study_concepts', 'mathematical')
g.edge('study_math', 'conditional_prob')
g.edge('study_conditional_prob', 'distributions')
g.edge('study_distributions', 'regressionML')
g.edge('study_regressionML', 'apply')

g.render('stats')

g
```
